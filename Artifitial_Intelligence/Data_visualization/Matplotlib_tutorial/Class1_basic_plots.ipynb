## Import

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

## Function plot

```python
x_1 = np.linspace(0, 5, 10)
print(x_1)
y_1 = x_1**2
print(y_1)
plt.plot(x_1, y_1)
plt.title('Days squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
plt.show()
```

## Multiple Plots
* arguments


```python
ax1 = plt.subplot(3,2,1)
plt.plot(x_1, y_1, 'red')
plt.title('Red')
plt.subplot(3,2,2)
plt.plot(x_1, y_1, 'blue')
plt.plot(x_1*2, y_1, 'Orange')
plt.title('Blue')
plt.subplot(3,2,3, sharex=ax1)
plt.plot(x_1, y_1, 'green')
plt.xticks(visible=False)
plt.title('Green')
plt.subplot(3,2,5)
plt.plot(x_1, y_1, 'black')
plt.title('Black')
```

## Using Figure Objects

### Creation of figure


### Adding axes



```python
fig_1 = plt.figure(figsize=(10,10), dpi=100)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('X')
axes_1.set_ylabel('Y')
axes_1.set_title('Days')
axes_1.plot(x_1, y_1, label='x/x2')
axes_1.plot(x_1 * 2, y_1, label='2x/x2', color='red')
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.1,0.1,0.45,0.45])
axes_2.text(5, 25, 'This is hierachial ploting')
axes_2.set_xlabel('X')
axes_2.set_ylabel('Y')
axes_2.set_title('Days')
axes_2.plot(x_1, y_1, 'Black')
```

## Subplots

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout()
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('Y')
axes_2[1].plot(x_1, y_1)
```

## Summary
* add_axes([x, y, width, height])
<br>
Create a figure first and add axes one by one on a desired postion with desired size.
<br>
fig_1 = plt.figure(figsize=(10,10), dpi=100)
<br>
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])

* subplots(figuresize, nrows, ncols)
<br>
Create a figure with subplots as many as you want. Returns are figure, and a list of axes
<br>
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)

* subplot(col_span + row_span + index)
<br>
Create a figure first and add subplot(axes) with desired span and index. No need to bind subplots to the figure.

```python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0, 0, 1, 1])
axes_3.plot(x_1, y_1, color='violet', alpha=0.5, lw=4, ls='-.', marker='o', markersize=10, markerfacecolor='y', markeredgecolor='black',
           markeredgewidth=5)
axes_3.set_xlim([0, 3])
axes_3.set_ylim([0, 3])
axes_3.grid(True, color='red', dashes=(10,2,5,2))
axes_3.set_facecolor('#F0F0F0')
```
